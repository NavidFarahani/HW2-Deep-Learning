```
#In the name of God
import numpy as np
def sigmoid(x):
    return 1/(1+np.exp(-x))

def gradient_L_reg(y,t,x_concatnated,the_lambda,w):
    return (y*(1-y)*(y-t))*x_concatnated   +   the_lambda*w

num_features=10
x=np.random.normal([1],size=(num_features))
x_concatnated=np.concatenate((x,np.array([1])))
t=sigmoid(0.5*x[0]+0.4*x[1]+0.1*np.mean(x[2:]))
w=np.random.normal(size=(1,num_features+1))# b and w are concatenated
####### Updating w and b
#w=w-lr*gradient(L_reg)
num_epochs=1
num_ite=1001
the_lambda=0
lr=0.1
#If the_lambda =0, almost always the error will be zero
#(But if initial assumtion of w be bad, we may get stuck in local minimum)
# Also if the_lambda=0.2, the error will be not zero,
# but the norm of w will be controlled

for epoch in range(num_epochs):
    for i in range(num_ite):
        z=np.dot(w,x_concatnated)
        y=sigmoid(z)
        grad=gradient_L_reg(y,t,x_concatnated,the_lambda,w)
        w=w-lr*grad
        if (i%20==0):
            print("\nthe_lambda is: "+str(the_lambda))
            print("Error for step "+str(i)+ " is : "+str(abs(t-y)))
            print("Norm2 of w for step "+str(i)+" is: "+str(np.sqrt(np.sum(abs(w**2)))))


```
